```python
import pandas as pd
```

```python
df= pd.read_csv("covid_toy.csv")
df.head(3)
```

```python
df.isnull().sum()  ###return total missing value in a dataframe
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer() ### it will file missing values

# if data
# numerical -----> fill data mean
#categorical ----> fill data most_frequent value
```

```python
df['fever'] =si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
df['has_covid'].value_counts()  ###it will return total frequency of each sun-category in coloumn
```

```python
df['gender'] =df['gender'].map({"Female":0 , "Male":1})
```

```python
df['cough']= df['cough'].map({"Mild":0,"Strong":1})
```

```python
df['city']=df['city'].map({
    "KOlkata":0,
    "Banglore":1,
    "Delhi":2,
    "Mumbai":3
})
```

```python
df['has_covid']=df['has_covid'].map({"no":0, "yes":1})
```

```python
df.sample(3)
```

```python
x=df.drop(columns=['has_covid'])   ### IOndependant data
```

```python
y=df['has_covid'] ### target/dependent data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x,
                                            y,
                                            test_size=0.2,
                                            random_state=42)
```
